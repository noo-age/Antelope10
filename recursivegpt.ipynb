```python
api_key = "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"
import openai
```

```python
def is_complete(text, api_key):
  prepend = "Determine whether the question or request contained at the top of the following query has been completed. Output 'y' or 'n' after the end of the query. {Beginning of query}"
  append = "{End of query}"
  response = openai.Completion.create(engine="text-davinci-003",
    prompt=prepend + text + append,
    max_tokens = 2000,
    api_key = api_key)
  if response.choices[0].text == "y":
    return True
  elif response.choices[0].text == "n":
    return False
  return is_complete(text,api_key)
  
def recursive_query(text, api_key):
  if is_complete(text,api_key):
    return text
  append = openai.Completion.create(engine="text-davinci-003",
    prompt=text,
    max_tokens=2000,
    api_key = api_key)
  text = text + append.choices[0].text
  return recursive_query(text,api_key)
```

```python
print(recursive_query("What is the largest prime number?", api_key))
```
